```python
def income_group(x):
    if x < 40000:
        return "Low"
    elif x < 80000:
        return "Medium"
    else:
        return "High"
df["income_group"] = df["person_income"].apply(income_group)        
```

```python
def loan_group(x):
    if x < 5000:
        return "Small"
    elif x < 15000:
        return "Medium"
    else:
        return "Large"

df['loan_size_group'] = df['loan_amnt'].apply(loan_group)
```

```python
def age_group(x):
    if x < 25:
        return "Young"
    elif x < 40:
        return "Mid"
    else:
        return "Senior"

df['age_group'] = df['person_age'].apply(age_group)
```

```python
def risk_flag(row):
    if row['person_income'] < 40000 and row['loan_amnt'] > 15000:
        return "High Risk"
    else:
        return "Low Risk"

df['risk_flag'] = df.apply(risk_flag, axis=1)
```

```python

df['risk_score'] = 0
df.loc[df['loan_size_group'] == 'Large', 'risk_score'] += 3
df.loc[df['loan_size_group'] == 'Medium', 'risk_score'] += 1


df.loc[df['income_group'] == 'Low', 'risk_score'] += 2


df.loc[df['age_group'] == 'Young', 'risk_score'] += 1

def assign_risk(score):
    if score <= 2:
        return 'Low Risk'
    elif score <= 4:
        return 'Medium Risk'
    else:
        return 'High Risk'

df['risk_level'] = df['risk_score'].apply(assign_risk)



df[['income_group','loan_size_group','age_group','risk_score','risk_level']].head()
```

```python
overall_default = df['loan_status'].mean()
print(overall_default)
```

```python
filtered_df = df[df['risk_level'] != 'High Risk']
new_default = filtered_df['loan_status'].mean()
print(new_default)
```

```python
print("Overall Default:", overall_default)
print("After Removing High Risk:", new_default)
```

```python
reduction = overall_default - new_default
print("Reduction in Default:", reduction)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create comparison dataframe
impact_df = pd.DataFrame({
    'Scenario': ['Before', 'After Removing High Risk'],
    'Default Rate': [overall_default, new_default]
})

# Plot
sns.barplot(
    data=impact_df,
    x='Scenario',
    y='Default Rate',
    palette='Set2'
)

plt.title("Impact of Removing High-Risk Customers")
plt.ylabel("Default Rate")
plt.show()
```

```python
threshold = df['risk_score'].quantile(0.9)
filtered_df_10 = df[df['risk_score'] < threshold]
new_default_10 = filtered_df_10['loan_status'].mean()
print(new_default_10)
```

```python
reduction_10 = overall_default - new_default_10
print(reduction_10)
```

```python
df_sorted = df.sort_values(by='risk_score', ascending=True)
df_sorted['cum_customers'] = range(1, len(df_sorted) + 1)
df_sorted['cum_approved_pct'] = df_sorted['cum_customers'] / len(df_sorted)
df_sorted['cum_default'] = df_sorted['loan_status'].cumsum()
df_sorted['cum_default_rate'] = df_sorted['cum_default'] / df_sorted['cum_customers']
```

```python
df_plot = df_sorted.iloc[100:]   

plt.plot(df_plot['cum_approved_pct'], df_plot['cum_default_rate'])
plt.xlabel("Percentage of Customers Approved")
plt.ylabel("Default Rate")
plt.title("Approval vs Default Trade-off Curve")
plt.show()
```
