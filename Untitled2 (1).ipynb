```python

df['person_emp_length'].fillna(df['person_emp_length'].median(), inplace=True)
df['loan_int_rate'].fillna(df['loan_int_rate'].median(), inplace=True)

df = df[df['person_age'] < 100]
df = df[df['person_income'] > 0]

df = df.drop_duplicates()
```

```python
def remove_outliers(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return df[(df[col] >= lower) & (df[col] <= upper)]

for col in ['person_income', 'loan_amnt', 'loan_int_rate']:
    df = remove_outliers(df, col)
```

```python
print(df.isnull().sum())
print("Duplicates:", df.duplicated().sum())
```

```python
kpis = {
    "Total Customers": len(df),
    "Default Rate (%)": round(df['loan_status'].mean()*100,2),
    "High Risk %": round((df['risk_segment']=='High Risk').mean()*100,2)
}

print(kpis)
```

```python
default_rate = df['loan_status'].mean()
print("Default Rate:", round(default_rate * 100, 2), "%")

df['loan_status'].value_counts().plot(kind='bar')
plt.title("Default vs Non-Default Count")
plt.show()
```

```python
sns.histplot(df['loan_income_ratio'], bins=50)
plt.title("Loan to Income Ratio Distribution")
plt.show()
```

```python
df['income_group'] = pd.cut(
    df['person_income'],
    bins=[0, 40000, 80000, 200000],
    labels=['Low', 'Medium', 'High']
)

df['loan_ratio_group'] = pd.cut(
    df['loan_income_ratio'],
    bins=[0, 0.2, 0.4, 1],
    labels=['Low', 'Medium', 'High']
)
```

```python
df['risk_segment'] = 'Low Risk'

df.loc[df['loan_income_ratio'] > 0.3, 'risk_segment'] = 'Medium Risk'

df.loc[
    (df['loan_income_ratio'] > 0.5) &
    (df['person_income'] < 40000),
    'risk_segment'
] = 'High Risk'
```

```python
df.groupby('risk_segment')['loan_status'].mean().plot(kind='bar')
plt.title("Default Rate by Risk Segment")
plt.show()
```

Analysis does not include external credit score or economic conditions.

####Executive Summary


This analysis identified high-risk borrowers based on income and loan ratio. A small segment of customers contributes disproportionately to defaults. By restricting high-risk approvals and focusing on safer segments, financial institutions can reduce default risk and improve loan performance.
